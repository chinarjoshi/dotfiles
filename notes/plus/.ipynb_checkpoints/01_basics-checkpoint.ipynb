#### CODING PROBLEM #1

A new night club has some wacky entrance rules. You must have been born between the years 1997-2003, and your age must be an odd number. In addition, your name cannot have more than 6 letters. Write a boolean function called `entrance` with a string argument 'name' and a number argument 'age'.

Note: There is a function 'len' that returns the integer length of a string.

Examples:

    >>> entrance('Chinar', 19)
    True
    >>> entrance('Doofenshmirtz', 24)
    False
    >>> entrance('Tim', 60)
    False

#### Solution:

````python
def entrance(name, age):
    return age > 18 and age < 25 and age % 2 == 1 and len(name) <= 6

entrance('Chinar', 19), entrance('Doofenshmirtz', 24), entrance('Tim', 60)
````

#### CODING PROBLEM #2

You're tired of solving single term power rule derivatives and integrals and want to make python do it. Recall the following formulas:

Derivative power rule: ax^n = n * ax^(n - 1)

Integral power rule: (x^(n+1))/(n + 1)

Write a function 'derivative' and a function 'integral', both with a floating point argument 'a', the coefficient, and an integer argument 'n', the exponent. The function should return a STRING in the format ax^n. The resulting 'a' should be capped at 3 decimals.

Examples:

    >>> derivative(4, 3)
    12x^2
    >>> integral(5, 6)
    .714x^7
    >>> derivative(1, 8)
    8x^7

#### Solution:

````python
def derivative(a, n):
    return f'{a * n}x^{n - 1}'

def integral(a, n):
    return f'{round(a / (n + 1), 3)}x^{n + 1}'

derivative(4, 3), integral(5, 6), derivative(1, 8)
````
